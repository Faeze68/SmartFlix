```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#  Load the Ratings and Movies Data
ratings_cols = ['user_id', 'item_id', 'rating', 'timestamp']
ratings = pd.read_csv("../data/u.data", sep="\t", names=ratings_cols, encoding="latin-1")

movie_cols = ['item_id', 'title', 'release_date', 'video_release', 'imdb_url']
movies = pd.read_csv("../data/u.item", sep="|", names=movie_cols, usecols=[0, 1], encoding="latin-1")

# Merge ratings with movie titles
ratings = ratings.merge(movies, on="item_id")

print(" Data loaded:")
print(ratings.head(), "\n")
```

```python
ratings_matrix = ratings.pivot_table(
    index="user_id",
    columns="title",
    values="rating"
)
print(f"Matrix shape: {ratings_matrix.shape}")
```

```python
# Fill Missing Values & Compute Similarity
ratings_filled = ratings_matrix.fillna(0)       # replace NaN with 0
similarity_matrix = cosine_similarity(ratings_filled.T)  # movies as rows
similarity_df = pd.DataFrame(
    similarity_matrix,
    index=ratings_filled.columns,
    columns=ratings_filled.columns
)

print("\n Similarity matrix ready!")
print(similarity_df.iloc[:5, :5])  # preview

movie_to_test = "Star Wars (1977)"
print(f"\nðŸŽ¬ Recommendations for '{movie_to_test}':")
print(recommend(movie_to_test))
```
